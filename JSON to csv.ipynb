```python
import json
import pandas as pd
```

<h1>For Reading from JSON file</h1>

```python
def read_from_json(path):
    fd = open(path,'r')
    r = fd.read()
    fd.close()

    record = json.loads(r)
    return record
```

```python
records = read_from_json("tweets.json")
len(records)
```

```python
tweet_ID_list = list(records.keys())
tweet_ID_list
```

```python
tweet_dict_list = []
for x in tweet_ID_list:
    tweet_dict = records[x]
    tweet_dict_list.append(tweet_dict)
tweet_dict_list
```

```python
tweet_author_list = []
tweet_text_list= []
for tweet in tweet_dict_list:
    tweet_author = tweet['tweet_author']
    tweet_author_list.append(tweet_author)
    tweet_text = tweet['tweet_text']
    tweet_text_list.append(tweet_text)
print(len(tweet_author_list))
print(len(tweet_text_list))
```

```python
df = pd.DataFrame(list(zip(tweet_author_list, tweet_text_list)),
               columns =[ 'Tweet_Author','Tweet_text'])
df
```

```python
df.to_csv('tweets.csv', encoding='utf-8', index=False)
```
